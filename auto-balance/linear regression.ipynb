```python
import pandas as pd
import json
from statistics import mean
```

```python
 #https://www.geeksforgeeks.org/read-json-file-using-python/
# data_source_file = "./data copy.json"
data_source_file = "./data-frame-friendly.json"

f = open (data_source_file, "r")
data = json.load(f)
df1 = pd.DataFrame.from_dict(data, orient='index')
df1=df1.reset_index()
df1
```

```python
df_test = pd.DataFrame.from_dict(data, orient='index')
df_test
```

```python
row_storage = []

for main_index,main_row in df1.iterrows(): 
#     print(main_index)
    meta_data = df1[['time','url','map','score_a','score_b']].iloc[main_index]

    teams = ["team_a","team_b"]

    for team_number in teams:
        df2 = pd.json_normalize(df1[team_number].iloc[main_index]) #take the first row 


        for index,row in df2.iterrows():
            new_row = pd.concat([meta_data, row])
            row_storage.append(new_row)


new_df = pd.DataFrame(row_storage)
new_df
```

```python
average_acs = new_df.loc[new_df['map']=="Ascent"]
average_acs
```

```python
# more than 1 condition : https://sparkbyexamples.com/pandas/pandas-loc-multiple-conditions/
# we can take the last 3 
def calc_player_acs_3(player_name,map_name): # we are taking the top
#     player_name = "brian"
    # def calc_player_average

    recent3_maps = new_df.loc[(new_df['player_name']==player_name)&(new_df['map']==map_name)]['average_combat_score'][-3:].tolist()
    
    recent3_games = new_df.loc[(new_df['player_name']==player_name)]['average_combat_score'][-3:].tolist()
    
    recent_6 = recent3_maps + recent3_games
    
#     print(recent3_maps,recent3_games)
#     print(recent_6)
    average_acs = mean(recent_6)

    
    return average_acs

test_one = calc_player_acs_3('andy','Ascent')
print(test_one)
```

```python
# we can take the last 3 
def calc_player_average(player_name): 
#     player_name = "brian"
    # def calc_player_average
    #dataFrame['Salary']>=100000) & (dataFrame['Age']< 40)
    kills_average = new_df.loc[new_df['player_name']==player_name]['kills'].tolist()
    kills_average = mean(kills_average)

    deaths_average = new_df.loc[new_df['player_name']==player_name]['deaths'].tolist()
    deaths_average = mean(deaths_average)

    assists_average = new_df.loc[new_df['player_name']==player_name]['assists'].tolist()
    assists_average = mean(assists_average)

    kill_deaths_average = new_df.loc[new_df['player_name']==player_name]['kill_deaths'].tolist()
    kill_deaths_average = mean(kill_deaths_average)

    kill_assist_survive_traded_average = new_df.loc[new_df['player_name']==player_name]['kill_assist_survive_traded'].tolist()
    kill_assist_survive_traded_average = mean(kill_assist_survive_traded_average)

    first_kills_average = new_df.loc[new_df['player_name']==player_name]['first_kills'].tolist()
    first_kills_average = mean(first_kills_average)

    first_deaths_average = new_df.loc[new_df['player_name']==player_name]['first_deaths'].tolist()
    first_deaths_average = mean(first_deaths_average)

    multi_kills_average = new_df.loc[new_df['player_name']==player_name]['multi_kills'].tolist()
    multi_kills_average = mean(multi_kills_average)

    econ_average = new_df.loc[new_df['player_name']==player_name]['econ'].tolist()
    econ_average = mean(econ_average)


    y_values = [kills_average,deaths_average,assists_average,kill_deaths_average,kill_assist_survive_traded_average,first_kills_average,first_deaths_average,multi_kills_average,econ_average]
    
    return y_values

test_one = calc_player_average('brian')
print(test_one)
```
